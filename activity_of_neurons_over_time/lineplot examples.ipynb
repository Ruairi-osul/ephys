```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('fmri')
```

```python
df.head()
```

```python
ax = sns.lineplot(x='timepoint', y='signal', data=df)
```

```python
ax = sns.lineplot(x='timepoint', y='signal', hue='event', data=df)
```

```python
ax = sns.lineplot(x='timepoint', y='signal', hue='event', style='event', 
                 data=df)
```

```python
ax = sns.lineplot(x='timepoint', y='signal', hue='region', style='event',
                 data=df)
```

```python
ax = sns.lineplot(x='timepoint', y='signal', hue='event', style='event',
                 markers=True, dashes=False, data=df)
```

```python
ax = sns.lineplot(x='timepoint', y='signal', hue='event', err_style='bars',
                 ci=60, data=df)
```

```python
ax = sns.lineplot(x='timepoint', y='signal', hue='event',
                 units='subject', estimator=None, lw=1, 
                  data=df[df['region']=='frontal'])
```

```python
ax = sns.lineplot(x='time', y='firing_rate', data=dots, 
                  style='choice')
```

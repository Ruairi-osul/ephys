```python
def _load_raw_data(path, num_channels=32):
    ''' given a path to a .dat file, loads np.memmap object of that file'''
    tmp = np.memmap(path, dtype=np.int16)  # adjust shape
    adjusted_len = int(len(tmp) / num_channels)
    
    return np.memmap(path, dtype=np.int16, shape=(adjusted_len, num_channels))
```
